```
#Import libraries
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Authentification Google
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

# Download du fichier
id = '1T0LqOHztZ1wpVl8O4u-qxcQ_kiwDehbb'
downloaded = drive.CreateFile({'id': '1T0LqOHztZ1wpVl8O4u-qxcQ_kiwDehbb'})
downloaded.GetContentFile('test_2.jpg')

#Read images for different blurring purposes
image = cv2.imread("test_2.jpg")
#Apply different thresholds and save in different variables
_, img1 = cv2.threshold(image, 50, 255, 0 )
_, img2 = cv2.threshold(image, 50, 255, 1 )
_, img3 = cv2.threshold(image, 50, 255, 2 )
_, img4 = cv2.threshold(image, 50, 255, 3 )
_, img5 = cv2.threshold(image, 50, 255, 4 )
#Show the different threshold images
plt.imshow(img1) #Prints Binary Image
plt.imshow(img2) #Prints Binary Inverted Image
plt.imshow(img3) #Prints Truncated Image
plt.imshow(img4) #Prints Threshold to Zero Image
plt.imshow(img5) #Prints Threshold to Zero Inverted Image
```
