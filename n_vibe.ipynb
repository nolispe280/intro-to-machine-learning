```
array_lon = []
array_lat = []
array_alt = []
array_label = []
array_mac_address = []
for index, row in df_pb.iterrows():  
    temp = row['pb_bl'][0:len(row['pb_bl'])-1]
    temp_val_array = temp[1:len(temp)-1].split(',')
    temp_lon = re.findall(r"[-+]?\d*\.\d+|\d+", temp_val_array[0])[0]
    array_lon.append(temp_lon)
    temp_lat = temp_val_array[1].split(':')[1]
    array_lat.append(temp_lat)
    temp_alt = temp_val_array[2].split(':')[1]
    array_alt.append(temp_alt)
    temp_label = temp_val_array[3].split(':')[1]
    array_label.append(temp_label)
    temp_mac_address = temp_val_array[4].strip("mac_address\:\"")
    array_mac_address.append(temp_mac_address)
```

```
def get_all_names_files(data):
  """Avoir la liste de tout les fichiers."""
  try :
    list_files = os.listdir(data)
  except OSError:
    print ("La lecture de tout les fichiers dans le dossier %s a échoué" % data)
  else:
    print ("Réussite de la lecture de tout les fichiers dans le dossier %s " % data)
  return list_files
```

```
list_parcourt = get_all_names_files("Mesures fixes")
```

```
data_lppi_txt = "Mesures fixes"

data_info = {}

for parcourt in list_parcourt:
  
  if len(parcourt) >= 34:
    temp_pm = ""
    temp_pers = "personne" 
    temp_it = "iteration"
    array_device = []
    array_rssi = []
    array_pm_lab = []
    array_pers = []
    array_it = []

    print(parcourt)

    temp_pers = temp_pers + parcourt[20]
    temp_it = temp_it + parcourt[20]

    if parcourt[5] == '1' :
      temp_pm = "A" + parcourt[11]
    elif parcourt[5] == '2' :
      temp_pm = "C" + parcourt[11]
    elif parcourt[5] == '3' :
      temp_pm = "D" + parcourt[11]
    elif parcourt[5] == '4' :
      temp_pm = "E" + parcourt[11]
    elif parcourt[5] == '5' :
      temp_pm = "G" + parcourt[11]
    elif parcourt[5] == '6' :
      temp_pm = "I" + parcourt[11]
    elif parcourt[5] == '7' :
      temp_pm = "K" + parcourt[11]
    elif parcourt[5] == '8' :
      temp_pm = "M" + parcourt[11]
    elif parcourt[5] == '9' :
      temp_pm = "N" + parcourt[11]
    else:
      temp_pm = "IMP"

    f = open(data_lppi_txt + '/' + parcourt, "r")
    info_lppi = f.readline().strip("delta t:500ms  nb points :50      ")
    info_lppi_array = info_lppi.split("NVibeRssiStorage")
    
    del info_lppi_array[0]

    print(parcourt)

    if len(info_lppi_array) >= 1:


      info_lppi_array[-1] = info_lppi_array[-1].strip('\]')

      for info in info_lppi_array:
        temp_info_device,temp_info_rssi = info.split("," , 1)
        temp_info_device = temp_info_device.strip('{device=')
        temp_info_rssi = temp_info_rssi.strip(' rssi=')
        temp_info_rssi = temp_info_rssi.rstrip('},')
        array_device.append(temp_info_device)
        array_rssi.append(temp_info_rssi)
        data_lppi = {'device' : array_device, 'rssi' : array_rssi}
        df_lppi  = pd.DataFrame(data= data_lppi)

        i = 2

        for i in range(df_lppi.shape[0]):
          array_pm_lab.append(temp_pm)
          array_pers.append(temp_pers)
          array_it.append(array_it)

        df_lppi['Personne'] = temp_pers
        df_lppi['Iteration'] = temp_it
        df_lppi['Label_pm'] = temp_pm
        df_lppi['lat'] = ""
        df_lppi['lon'] = ""

        for index_lppi, row_lppi in df_lppi.iterrows():

          for index_pm, row_pm in df_pm.iterrows() :

            if row_lppi['Label_pm'] == row_pm['label'] :

              row_lppi['lat'] = row_pm['lat']
              row_lppi['lon'] = row_pm['lon']


  data_info[parcourt] = df_lppi


  
    




  

```
