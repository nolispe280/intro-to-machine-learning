```
array_pm = {}

label_array = []
lat_array = []
lon_array = []

for index, row in df_pm.iterrows():  
    print(row['pm_rssi'])
    x = row['pm_rssi'].split(":")
    array_pm[x[0].rstrip()] = [float(i) for i in x[1].lstrip().split(",")]
    temp = [float(i) for i in x[1].lstrip().split(",")]
    label_array.append(x[0].rstrip())
    lat_array.append(temp[0])
    lon_array.append(temp[1])
```

```
data_pm = {'label' : label_array, 'lat' : lat_array, 'lon' : lon_array}
df_pm = pd.DataFrame(data= data_pm)
```

```
array_lon = []
array_lat = []
array_alt = []
array_label = []
array_mac_address = []
for index, row in df_pb.iterrows():  
    temp = row['pb_bl'][0:len(row['pb_bl'])-1]
    temp_val_array = temp[1:len(temp)-1].split(',')
    temp_lon = re.findall(r"[-+]?\d*\.\d+|\d+", temp_val_array[0])[0]
    temp_lat = temp_val_array[1].split(':')[1]
    temp_alt = temp_val_array[2].split(':')[1]
    temp_label = temp_val_array[3].split(':')[1]
    temp_mac_address = temp_val_array[4].split(':')[1:-1]
    print(temp_lon,"-",temp_lat,"-",temp_alt,"-",temp_label,"-",temp_mac_address)
```
